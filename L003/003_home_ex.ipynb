```python
#перечень всех документов

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

#перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть 
#и в directories)

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
#задание №1
#Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
#Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
#Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
#Пункт 4. Пользователь по команде "ads" может добавить новую полку
#Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
#задание #2
#Пункт 1. Пользователь по команде "ad" может добавить новый документ в данные
#Пункт 2. Пользователь по команде "d" может удалить документ из данных
#Пункт 3. Пользователь по команде "m" может переместить документ с полки на полку
```

```python
#проверить наличие документа и вернуть имя владельца
def doc_in_documents(doc_num):
    '''
    проверяет наличие документа в списке документов по его номеру
    выводит имя владельца документа в случае удачи или None в случае отсутствия документа
    '''
    for doc in documents:
        if doc['number'] == doc_num:
            return doc['name']
    return None
#на какой полке документ
def doc_in_directories(doc_num):
    '''
    находит документ в шкафу по его номеру
    выводит номер полки в случае удачи или None в случае отсутствия документа
    '''
    for directory in directories:
        if directories[directory].count(doc_num) != 0:
            return directory
    return None
#добавить полку
def dir_add(dir_num):
    '''
    добавляет пустую полку в шкаф
    выводит перечень полок и, в случае удачи, None, а в случае наличия полки с таким номером - True
    '''
    for directory in directories:
        if directory == dir_num:          
            return True, list(directories.keys())
    directories[dir_num] = []
    return None, list(directories.keys())
#удаление пустой полки
def dir_del(dir_num):
    '''
    удаляет пустую полку из шкафа
    выводит перечень полок и, в случае удачи, True, в случае, если на полке есть документы - False,
    в случае отсутствия в шкафу полки с таким номером - None
    '''
    for directory in directories:
        if directory == dir_num:
            if len(directories[directory]) == 0:
                del(directories[directory])
                return True, list(directories.keys())
            else:
                return False, list(directories.keys())
    return None, list(directories.keys())
#добавление нового документа в список и на полку
def doc_in_list():
    '''
    добавляет новый документ в список документов
    возвращает в случае удачи True, в случае, если документ уже существует - None,
    в случае отсутствия в шкафу полки с таким номером - False
    '''
    doc_num = input('Введите номер документа:')
    if doc_in_documents(doc_num) == None:
        dir_num = input('Введите номер полки: ')
        for directory in directories:
            if directory == dir_num:
                doc_type = input('Введите тип документа:')
                doc_owner = input('Введите имя владельца документа:')
                documents.append({'type' : doc_type, 'number' : doc_num, 'name' : doc_owner})
                directories[directory].append(doc_num)
                return True
        return False               
    else:
        return None
#текущий список документов
def doc_list():
    '''
    выводит текущий список документов
    ничего не возвращает
    '''
    print('Текущий список документов:')
    for doc in documents:
        doc_num = doc['number']
        doc_type = doc['type']
        doc_name = doc['name']
        print(f'№: {doc_num}, тип: {doc_type}, владелец: {doc_name}, полка хранения: {doc_in_directories(doc_num)}')
#удаляет документ из данных
def doc_del():
    '''
    удаляет документ из данных
    возращает в случае удачи, True, а в случае отсутствия документа с таким номером - None,
    в случае наличия документа, но отсутствия его на полке - False
    '''
    doc_num = input('Введите номер документа:')
    for doc in documents:
        if doc['number'] == doc_num:
            directory = doc_in_directories(doc_num)
            if  directory != None:
                documents.remove(doc)
                directories[directory].remove(doc_num)
                return True
            else:
                return False
    return None
#перемещает документ между полками
def doc_move():
    '''
    перемещает документ между полками
    возвращает в случае удачи True, в случае, если документ не существует - None,
    в случае отсутствия в шкафу полки с таким номером - False
    '''
    doc_num = input('Введите номер документа:')
    if doc_in_documents(doc_num) != None:
        dir_num = input('Введите номер полки: ')
        for directory in directories:
            if directories[directory].count(doc_num) != 0:
                dir_doc_for_del = directory
            if directory == dir_num:
                directories[directory].append(doc_num)
                directories[dir_doc_for_del].remove(doc_num)
                return True
        return False               
    else:
        return None
```

```python
#основной цикл - цикл запроса команд
cmd = ''
while cmd != 'q':
    cmd = input('Введите команду: ')
    if cmd == 'p':
        ret = doc_in_documents(input('Введите номер документа: '))
        if ret != None:
            print(f'Владелец документа: {ret}')
        else:
            print('Документ не найден в базе.')
    elif cmd == 's':
        ret = doc_in_directories(input('Введите номер документа: '))
        if ret != None:
            print(f'Документ хранится на полке: {ret}')
        else:
            print('Документ не найден в базе.')
    elif cmd == 'l':
        doc_list()
    elif cmd == 'ads':
        ret, dir_list = dir_add(input('Введите номер полки: '))
        if ret != None:
            print('Такая полка уже существует. Текущий перечень полок: ', ', '.join(dir_list))
        else:
            print('Полка добавлена. Текущий перечень полок: ', ', '.join(dir_list))
    elif cmd == 'ds':    
        ret, dir_list = dir_del(input('Введите номер полки: '))
        if ret == True:
            print('Полка удалена. Текущий перечень полок: ', ', '.join(dir_list))
        elif ret == False:
            print('На полке есть документы. Удалите их перед удалением полки. Текущий перечень полок: ', ', '.join(dir_list))
        else:
            print('Такая полка не существует. Текущий перечень полок: ', ', '.join(dir_list))
    elif cmd == 'ad':
        ret = doc_in_list()
        if ret == None:
            print('Такой документ уже существует.')
        elif ret == False:
            print('Такой полки не существует. Добавьте полку командой ads.')
            doc_list()
        else:
            print('Документ добавлен.')
            doc_list()
    elif cmd == 'd':
        ret = doc_del()
        if ret == None:
            print('Такой документ не существует.')
            doc_list()
        elif ret == False:
            print('Такой документ существует, но не выложен на полку. Наказать секретаршу командой punish. ')
        else:
            print('Документ удален.')
            doc_list()
    elif cmd == 'm':
        ret = doc_move()
        if ret == None:
            print('Такой документ не существует.')
            doc_list()
        elif ret == False:
            print('Такой полки не существует.Текущий перечень полок: ',  ', '.join(list(directories.keys())))
        else:
            print('Документ перемещен.')
            doc_list()
    else:
        print(f'Неизвестная команда "{cmd}".')
print('Сеанс завершен.')
```
