````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
data = pd.read_csv('vgsales.csv')
data.info()
data.head()
````

````python
data.describe()
````

````python
#1) Как критики относятся к спортивным играм?

#H0 - средняя экспертная оценка игр спортивной тематики <= 80 баллов
#H1 - средняя экспертная оценка игр спортивной тематики > 80 баллов
#уровень значимости определим 0.05
````

````python
sports = data[data['Genre']=='Sports'].copy()
sports.dropna(subset = ['Critic_Score'], axis=0, inplace=True)
````

````python
import scipy.stats as stats

alpha = 0.05
t, p= stats.ttest_1samp(sports['Critic_Score'], 80, alternative='greater')

print(sports['Critic_Score'].mean())
print(f't: {t}, p-value: {p}')

if p < alpha: 
    print('Отвергаем нулевую гипотезу, среднее больше 80 (спортивные игры нравятся)')
else:
    print('Не отвергаем нулевую нулевую гипотезу, среднее меньше или равно 80 (спортивные игры НЕ нравятся)')
````

````python
#2) Критикам нравятся больше игры на PC или на PS4?

#H0 - средняя экспертная оценка игр на PC и PS4 не отличается
#H1 - средняя экспертная оценка игр на PC и PS4 отличается
#уровень значимости определим 0.05
````

````python
pc = data[data['Platform']=='PC'].copy()
ps4 = data[data['Platform']=='PS4'].copy()
pc.dropna(subset = ['Critic_Score'], axis=0, inplace=True)
ps4.dropna(subset = ['Critic_Score'], axis=0, inplace=True)
````

````python
t, p = stats.ttest_ind(pc['Critic_Score'], ps4['Critic_Score'], equal_var=False)
print(f't: {t}, p-value: {p}')
print(f'Средние значения PC: {pc["Critic_Score"].mean()}, PS4: {ps4["Critic_Score"].mean()}')

if (p < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к играм на PC и PS4')
    if pc["Critic_Score"].mean() > ps4["Critic_Score"].mean():
        print('Критикам больше нравятся игры на PC')
    elif pc["Critic_Score"].mean() < ps4["Critic_Score"].mean():
        print('Критикам больше нравятся игры на PS4')
else:
    print('Не отвергаем нулевую гипотезу')
````

````python
#3) Критикам больше нравятся стрелялки или стратегии?

#H0 - средняя экспертная оценка стрелялок и стратегий не отличается
#H1 - средняя экспертная оценка стрелялок и стратегий отличается
#уровень значимости определим 0.05
````

````python
shooter = data[data['Genre']=='Shooter'].copy()
strategy = data[data['Genre']=='Strategy'].copy()
shooter.dropna(subset = ['Critic_Score'], axis=0, inplace=True)
strategy.dropna(subset = ['Critic_Score'], axis=0, inplace=True)
````

````python
t, p = stats.ttest_ind(shooter['Critic_Score'], strategy['Critic_Score'], equal_var=False)
print(f't: {t}, p-value: {p}')
print(f'Средние значения "стрелялки": {shooter["Critic_Score"].mean()}, "стратегии": {strategy["Critic_Score"].mean()}')

if (p < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к стрелялкам и стратегиям')
    if shooter["Critic_Score"].mean() > strategy["Critic_Score"].mean():
        print('Критикам больше нравятся стрелялки')
    elif shooter["Critic_Score"].mean() < strategy["Critic_Score"].mean():
        print('Критикам больше нравятся стратегии')
else:
    print('Не отвергаем нулевую гипотезу')
````
