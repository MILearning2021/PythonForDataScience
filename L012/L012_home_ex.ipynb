```python
def source_type(row): 
    if row['traffic_source'] in ['yandex', 'google']:
        return 'organic'
    elif (row['traffic_source'] in ['paid', 'email']) and row['region'] == 'Russia':
        return 'ad'
    elif (row['traffic_source'] in ['paid', 'email']) and row['region'] != 'Russia':
        return 'other'
    else:
        return row['traffic_source']

log['source_type'] = log.apply(source_type, axis = 1)
```

```python
# Задание №3

ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
#получаю массив пользователь - количество оценок
ratings_count = ratings.groupby('userId').agg({'rating':'count'})
```

```python
%%time
#время жизни v1
ttl = 0
for userId in ratings_count[ratings_count['rating'] > 100].index: #пользователи с >, чем 100 оценок
    ttl += ratings[ratings['userId'] == userId]['timestamp'].max() \
                    - ratings[ratings['userId'] == userId]['timestamp'].min()

print(f'Среднее время "жизни" пользователя: \
{round((ttl / len(ratings_count[ratings_count["rating"] > 100].index))/ 86400, 2)} суток')
```

```python
%%time
#v2
ratings_count_tmin = ratings.groupby('userId').agg({'rating':'count', 'timestamp':'min'})
ratings_count_tmax = ratings.groupby('userId').agg({'timestamp':'max'})
joined = ratings_count_tmin.join(ratings_count_tmax, lsuffix = '_min', rsuffix = '_max')
joined['delta_time'] = joined['timestamp_max'] - joined['timestamp_min']
joined_100 = joined[joined['rating']>100].groupby('userId').agg({'delta_time' : 'sum'})
summa = int(joined_100.sum())
dlinna = joined_100.value_counts().sum()
print(f'Среднее время "жизни" пользователя: {round(summa / dlinna / 86400, 2)} суток')
```
